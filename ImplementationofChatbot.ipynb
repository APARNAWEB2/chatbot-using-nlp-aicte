# vectorizer
## CountVectorizer: Transforms the text into a matrix of token counts (word frequencies), capturing the basic frequency of each word within the documents. This method is useful for analyzing the overall presence of specific terms in the data.

## TfidfVectorizer: Converts text into a matrix that reflects the importance of words within a document relative to their frequency across all documents. This technique helps emphasize meaningful and distinctive terms, allowing the model to focus on keywords that differentiate the data.

##  Classifier Selection:-
#### -> Logistic Regression
#### ->Multinomial Naive Bayes

```python
#create the vectorizer and classifier
 
vectorizer=TfidfVectorizer()
clf1=LogisticRegression(random_state=0,max_iter=10000)
#preprocessing the data
tags1=[]
patterns1=[]
for intent in intents:
    for pattern in intent['patterns']:
        tags1.append(intent['tag'])
        patterns1.append(pattern)
#training the model
x1=vectorizer.fit_transform(patterns1)
y1=tags1
clf1.fit(x1,y1)
```

```python
#create the vectorizer and classifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import numpy as  np
vectorizer=CountVectorizer()
np.random.seed(0)
clf2= MultinomialNB()

#preprocessing the data
tags2=[]
patterns2=[]
for intent in intents:
    for pattern in intent['patterns']:
        tags2.append(intent['tag'])
        patterns2.append(pattern)
#training the model
x2=vectorizer.fit_transform(patterns2)
y2=tags2
clf2.fit(x2,y2)
```

```python
from sklearn.metrics import accuracy_score

# Predictions
y_pred1 = clf1.predict(x1)
y_pred2 = clf2.predict(x2)

# Accuracy
accuracy1 = accuracy_score(y1, y_pred1)
accuracy2 = accuracy_score(y2, y_pred2)

print("Logistic Regression with TF-IDF Vectorizer Accuracy:", accuracy1)
print("Multinomial Naive Bayes with Count Vectorizer Accuracy:", accuracy2)
```

```python
#python fuction to chat with the chabot
def chatbot(input_text):
    input_text=vectorizer.transform([input_text])
    tag=clf2.predict(input_text)[0]
    for intent in intents:
        if intent['tag']==tag:
            response=random.choice(intent['responses'])
            return response

        
```
